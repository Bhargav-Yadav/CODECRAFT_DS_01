```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
try:
    df = pd.read_csv('Population.csv', skiprows=4)
except Exception as e:
    print(e)
    try:
        df = pd.read_csv('Population.csv', sep='\t', skiprows=4)
    except Exception as e:
        print(e)
        try:
            df = pd.read_csv('Population.csv', sep=';', skiprows=4)
        except Exception as e:
            print(e)

df.head()
```

```python
df_final = df.drop(['Country Code', 'Indicator Name', 'Indicator Code'], axis = 1)
```

```python
chosen_country = 'India'
country_data = df_final[df_final['Country Name'] == chosen_country].drop('Unnamed: 69', axis=1)

years = country_data.columns[1:].astype(int)
population = country_data.iloc[:, 1:].values.flatten()

plt.figure(figsize=(5, 3))
plt.bar(years, population, color='skyblue')
plt.xlabel('Year')
plt.ylabel('Population')
plt.title(f'Population of {chosen_country} over time since 1960')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(population, bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.title(f'Histogram of Population for {chosen_country}')
plt.grid(axis='y', alpha=0.75)
plt.show()
```
